### Importing the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as s
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import *
from sklearn.metrics import confusion_matrix
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
%matplotlib inline
```

### Creating the dataset

```python
dataset = pd.read_csv("/content/gdrive/My Drive/Breast Cancer/data.csv")
dataset_new = pd.DataFrame(preprocessing.scale(dataset.ix[:,2:32]))
dataset_new.columns = list(dataset.ix[:,2:32].columns)
dataset_new['diagnosis'] = dataset['diagnosis']
X = dataset_new.ix[:,0:30]
y = dataset_new['diagnosis']
class_names = list(y.unique())
```

### Splitting the dataset into test and train

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

### Using the Support Vector Machines algorithm and representing its confusion matrix

```python
svc = SVC(kernel = 'linear',C=0.1, gamma=0.1, probability = True)
svc.fit(X,y)
svc_pred = svc.fit(X_train, y_train).predict(X_test)
svc_confusion = confusion_matrix(y_test, svc_pred)
print(svc.score(X_train,y_train), svc.score(X_test, y_test))
svca = svc.score(X_test, y_test)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(svc_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title('Support Vector Machine Confusion Matrix')
```

### Using the Logistic Regression algorithm and representing its Confusion Matrix

```python
lr = LogisticRegression(penalty = 'l1', dual = False, C=0.5)
lr_pred = lr.fit(X_train, y_train).predict(X_test)
print(lr.score(X_train,y_train), lr.score(X_test, y_test))
lra = lr.score(X_test, y_test)
lr_confusion = confusion_matrix(y_test, lr_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(lr_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title('Logistic Regression Confusion Matrix')
```

### Using the Multi-Layer Perceptron algorithm and representing its confusion matrix

```python
clf = MLPClassifier(solver='lbfgs', alpha=5, hidden_layer_sizes=(500,), random_state=10, )
mlp_fit = clf.fit(X_train, y_train)
mlp_pred = mlp_fit.predict(X_test)
print(mlp_fit.score(X_train, y_train), mlp_fit.score(X_test, y_test))
mlpa = mlp_fit.score(X_test, y_test)
mlp_confusion = confusion_matrix(y_test, mlp_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(mlp_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title('Multi-Layer Perceptron Confusion Matrix')
```

### Using the k-Nearest Neighbor classifier and representing its confusion matrix

```python
knr = KNeighborsClassifier(n_neighbors=9).fit(X_train, y_train)
print(knr.score(X_train, y_train), knr.score(X_test, y_test))
knra = knr.score(X_test, y_test)
knr_pred = knr.predict(X_test)
knr_confusion = confusion_matrix(y_test, knr_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(knr_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title('K Nearest Neighbors Confusion Matrix')
```

### Plotting the bar graph of accuracies of all the machine learning algorithms used earlier

```python
names=['LR','KNN','SVM','MLP']
models=[]
models.extend([lra,knra,svca,mlpa])
df=pd.DataFrame(models,index=names)
ax=df.plot.bar(legend=False)
rects = ax.patches
labels = [lra,knra,svca,mlpa]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height+0.05, label, ha='center')
ax.set_xlabel("Classifiers")
ax.set_ylabel("Accuracy")
plt.show()
```

## Since Multi-Layer Perceptron gave the best accuracy, we try to ensemble that algorithm with other machine learning algorithms to check if we get a better prediction accuracy or not

### Ensembling the Logistic Regression and Multi-Layer Perceptron Classifier, and representing its confusion matrix

```python
fusionlm = VotingClassifier(estimators=[('LR', lr), ('MLP', clf)], 
                       voting='soft', weights=[2,1]).fit(X_train,y_train)
print(fusionlm.score(X_train,y_train), fusionlm.score(X_test,y_test))
lma = fusionlm.score(X_test, y_test)
fusionlm_pred = fusionlm.predict(X_test)
fusionlm_confusion = confusion_matrix(y_test, fusionlm_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(fusionlm_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title(' Fusion LR and MLP Confusion Matrix')
```

### Ensembling the Multi-Layer Perceptron and k-Nearest Neighbor Classifier, and representing its confusion matrix

```python
fusionkl = VotingClassifier(estimators=[('KNN', knr), ('MLP', clf)], 
                       voting='soft', weights=[2,1]).fit(X_train,y_train) 
print(fusionkl.score(X_train,y_train), fusionkl.score(X_test,y_test))
kla = fusionkl.score(X_test, y_test)
fusionkl_pred = fusionkl.predict(X_test)
fusionkl_confusion = confusion_matrix(y_test, fusionkl_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(fusionkl_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title(' Fusion MLP and KNN Confusion Matrix')
```

### Ensembling the Multi-Layer Perceptron and Support Vector Machine Classifiers, and representing its confusion matrix

```python
fusionls = VotingClassifier(estimators=[('MLP', clf), ('SVC', svc)], 
                       voting='soft', weights=[1,2]).fit(X_train,y_train)
print(fusionls.score(X_train,y_train), fusionls.score(X_test,y_test))
lsa = fusionls.score(X_test, y_test)
fusionls_pred = fusionls.predict(X_test)
fusionls_confusion = confusion_matrix(y_test, fusionls_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(fusionls_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title(' Fusion MLP and SVM Confusion Matrix')
```

### Plotting the bar graph of accuracies of all the Ensemble classifiers used earlier

```python
names=['MLP AND SVM','LR AND MLP','MLP AND KNN']
models=[]
models.extend([lsa,lma,kla])
df=pd.DataFrame(models,index=names)
ax=df.plot.bar(legend=False)
rects = ax.patches
labels = [lsa,lma,kla]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height+0.05, label, ha='center')
ax.set_xlabel("Classifiers")
ax.set_ylabel("Accuracy")
plt.show()
```

### Ensembling the Multi-Layer Perceptron and Logistic Regression and k-Nearest Neighbors Classifiers, and representing its confusion matrix

```python
fusionklm = VotingClassifier(estimators=[('KNN', knr), ('LR', lr), ('MLP', clf)], 
                       voting='soft', weights=[1,2,3]).fit(X_train,y_train)
print(fusionklm.score(X_train,y_train), fusionklm.score(X_test,y_test))
klma = fusionklm.score(X_test, y_test)
fusionklm_pred = fusionklm.predict(X_test)
fusionklm_confusion = confusion_matrix(y_test, fusionklm_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(fusionklm_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title(' Fusion LR, MLP and KNN Confusion Matrix')
```

### Ensembling the Multi-Layer Perceptron and Logistic Regression and Support Vector Machine Classifiers, and representing its confusion matrix

```python
fusionlms = VotingClassifier(estimators=[('SVC', svc), ('LR', lr), ('MLP', clf)], 
                       voting='soft', weights=[1,2,3]).fit(X_train,y_train)
print(fusionlms.score(X_train,y_train), fusionlms.score(X_test,y_test))
slma = fusionlms.score(X_test, y_test)
fusionlms_pred = fusionlms.predict(X_test)
fusionlms_confusion = confusion_matrix(y_test, fusionlms_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(fusionlms_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title(' Fusion LR, SVC and MLP Confusion Matrix')
```

### Ensembling the Multi-Layer Perceptron, Support Vector Machine and k-Nearest Neighbors Classifiers, and representing its confusion matrix

```python
fusionkls = VotingClassifier(estimators=[('KNN', knr), ('MLP', clf), ('SVC', svc)], 
                       voting='soft', weights=[1,2,2]).fit(X_train,y_train)
print(fusionkls.score(X_train,y_train), fusionkls.score(X_test,y_test))
klsa = fusionkls.score(X_test, y_test)
fusionkls_pred = fusionkls.predict(X_test)
fusionkls_confusion = confusion_matrix(y_test, fusionkls_pred)
plt.rcParams['figure.figsize']=(14,8)
ax = plt.axes()
s.heatmap(fusionkls_confusion, annot=True, fmt='d', ax=ax, cmap="YlGnBu", annot_kws={"size": 30})
ax.set_title(' Fusion LR, SVC and KNN Confusion Matrix')
```

### Plotting the accuracy bar graphs of the three ensemble classifiers shown earlier

```python
names=['LR, MLP AND SVM','LR, KNN AND MLP','MLP, SVM AND KNN']
models=[]
models.extend([slma,klma,klsa])
df=pd.DataFrame(models,index=names)
ax=df.plot.bar(legend=False)
rects = ax.patches
labels = [slma,klma,klsa]
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height+0.05, label, ha='center')
ax.set_xlabel("Classifiers")
ax.set_ylabel("Accuracy")
plt.show()
```

### From the accuracy bar graphs of stand alone classifiers and ensemble classifers, we can conclude that the Multi-Layer Perceptron algorithm gave us the best accuracy in terms of prediction
